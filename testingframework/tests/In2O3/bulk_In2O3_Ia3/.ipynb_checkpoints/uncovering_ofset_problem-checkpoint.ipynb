````python
names = ['0_isolated_atom.xyz', '1_isolated_atom.xyz', '2_isolated_atom.xyz', '3_Slab_5f5e8a8ad507dedf0ee10ce9.xyz']
for i in range(4):
    ati_qe = seltraj[i]
    ati_mEspresso = ase.io.read(f'/data/lls34/womble0/runs/0000/0090_Run_210607_19:19:48/QE_recalculate_{i}.xyz')
    ati_Espresso = ase.io.read(os.path.join('/data/lls34/womble0/runs/0000/0094_Run_210608_13:30:00', names[i]))
    #ati_gap = ase.io.read(f'/data/lls34/womble0/runs/0000/0090_Run_210607_19:19:48/QE_recalculate_{i}.xyz')

    print(f'{i}th ')
    print('  Diff (train vs mEspresso): ',(ati_mEspresso.get_potential_energy() - ati_qe.get_potential_energy())/ati_qe.get_global_number_of_atoms())
    print('  Diff (train vs Espresso): ',(ati_mEspresso.get_potential_energy() - ati_Espresso.get_potential_energy())/ati_qe.get_global_number_of_atoms())
    print('  Training set: ',(ati_qe.get_potential_energy())/ati_qe.get_global_number_of_atoms())    
    print('  mEspresso: ',( ati_mEspresso.get_potential_energy())/ati_qe.get_global_number_of_atoms())     
    print('  Espresso: ',(ati_Espresso.get_potential_energy() )/ati_qe.get_global_number_of_atoms())            
````
